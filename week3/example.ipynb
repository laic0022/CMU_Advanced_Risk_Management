```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras import layers, Model
import yfinance as yf
import matplotlib.pyplot as plt

# Define the TransformerModel class
class TransformerModel(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, seq_length, num_transformer_blocks, dropout_rate=0.1):
        super(TransformerModel, self).__init__()
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.ff_dim = ff_dim
        self.seq_length = seq_length
        self.num_transformer_blocks = num_transformer_blocks

        self.positional_encoding = self.positional_encoding(seq_length)

        self.transformer_blocks = [
            self.transformer_block(embed_dim, num_heads, ff_dim, dropout_rate)
            for _ in range(num_transformer_blocks)
        ]

        self.dropout_1 = layers.Dropout(dropout_rate)
        self.flatten = layers.Flatten()
        self.dense = layers.Dense(1)

    def call(self, inputs, training=True):
        seq_length = inputs.shape[1]
        word_emb = inputs

        # Tile the positional encoding to match the batch size
        pos_encoding_tiled = tf.tile(self.positional_encoding, [tf.shape(inputs)[0], 1, 1])

        # Add positional encoding to word embeddings
        word_emb += pos_encoding_tiled

        x = self.dropout_1(word_emb, training=training)
        for i in range(self.num_transformer_blocks):
            x = self.transformer_blocks[i](x, training=training)
        x = self.flatten(x)
        output = self.dense(x)
        return output

    def transformer_block(self, embed_dim, num_heads, ff_dim, dropout_rate):
        inputs = layers.Input(shape=(None, embed_dim))

        # Self-attention layer
        attention = layers.MultiHeadAttention(
            num_heads=num_heads, key_dim=embed_dim, dropout=dropout_rate
        )(inputs, inputs)
        attention = layers.Dropout(dropout_rate)(attention)
        attention = layers.LayerNormalization(epsilon=1e-6)(inputs + attention)

        # Feed-forward layer
        outputs = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(attention)
        outputs = layers.Dropout(dropout_rate)(outputs)
        outputs = layers.Conv1D(filters=embed_dim, kernel_size=1)(outputs)
        outputs = layers.Dropout(dropout_rate)(outputs)
        outputs = layers.LayerNormalization(epsilon=1e-6)(attention + outputs)

        return Model(inputs=inputs, outputs=outputs)

    def positional_encoding(self, seq_length):
        pos = tf.cast(tf.range(seq_length)[:, tf.newaxis], dtype=tf.float32)
        i = tf.cast(tf.range(self.embed_dim)[tf.newaxis, :], dtype=tf.float32)
        angle_rads = pos / tf.pow(10000, 2 * (i // 2) / tf.cast(self.embed_dim, tf.float32))
        angle_rads = tf.where(tf.math.equal(i % 2, 0), tf.sin(angle_rads), tf.cos(angle_rads))
        pos_encoding = angle_rads[tf.newaxis, ...]
        return pos_encoding


# Function to fetch historical stock prices for AAPL
def fetch_stock_data(stock_symbol, start_date, end_date):
    stock_data = yf.download(stock_symbol, start=start_date, end=end_date)
    return stock_data

# Function to generate training data from historical stock prices
def generate_data(stock_data, seq_length):
    closing_prices = stock_data['Close'].values.reshape(-1, 1)
    scaler = MinMaxScaler()
    scaled_prices = scaler.fit_transform(closing_prices)
    X, y = [], []
    for i in range(len(scaled_prices) - seq_length):
        X.append(scaled_prices[i:i+seq_length])
        y.append(scaled_prices[i+seq_length])
    X = np.array(X)
    y = np.array(y)
    return X, y, scaler

# Define training step
@tf.function
def train_step(inputs, targets):
    with tf.GradientTape() as tape:
        predictions = model(inputs, training=True)
        loss = loss_fn(targets, predictions)
    gradients = tape.gradient(loss, model.trainable_variables)
    optimizer.apply_gradients(zip(gradients, model.trainable_variables))
    return loss

# Fetch historical data for AAPL
stock_symbol = 'AAPL'
start_date = '2012-01-01'
end_date = '2022-01-01'
stock_data = fetch_stock_data(stock_symbol, start_date, end_date)

# Hyperparameters
embed_dim = 32
num_heads = 2
ff_dim = 32
seq_length = 20
num_transformer_blocks = 2
dropout_rate = 0.1
learning_rate = 0.001
batch_size = 64
epochs = 10

# Generate training data
X_train, y_train, scaler = generate_data(stock_data, seq_length)

# Initialize model
model = TransformerModel(
    embed_dim=embed_dim,
    num_heads=num_heads,
    ff_dim=ff_dim,
    seq_length=seq_length,
    num_transformer_blocks=num_transformer_blocks,
    dropout_rate=dropout_rate
)

# Loss function and optimizer
loss_fn = tf.losses.MeanSquaredError()
optimizer = tf.optimizers.Adam(learning_rate)

# Training loop
num_batches = len(X_train) // batch_size
for epoch in range(epochs):
    total_loss = 0.0
    for i in range(num_batches):
        start_idx = i * batch_size
        end_idx = (i + 1) * batch_size
        batch_X, batch_y = X_train[start_idx:end_idx], y_train[start_idx:end_idx]
        loss = train_step(batch_X, batch_y)
        total_loss += loss
    average_loss = total_loss / num_batches
    print(f"Epoch {epoch+1}/{epochs}, Loss: {average_loss}")

# Prediction
next_days = 10
predicted_prices = []
last_sequence = X_train[-1]
for _ in range(next_days):
    # Predict the next price
    predicted_price = model(last_sequence.reshape(1, seq_length, 1))
    predicted_prices.append(predicted_price.numpy()[0][0])
    # Update the sequence by removing the first element and adding the predicted price
    last_sequence = np.concatenate((last_sequence[1:], [[predicted_price.numpy()[0][0]]]))

# Inverse transform the predicted prices
predicted_prices = scaler.inverse_transform(np.array(predicted_prices).reshape(-1, 1))

# Plot actual vs. predicted prices for the next 10 days
plt.figure(figsize=(10, 6))
plt.plot(stock_data.index[-20:], stock_data['Close'].values[-20:], label='Actual Prices', color='blue')  # Plot 20 days of actual prices
plt.plot(stock_data.index[-10:], predicted_prices, label='Predicted Prices', color='red')  # Plot predicted prices for the next 10 days
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Actual vs. Predicted Prices for the Next 10 Days')
plt.legend()
plt.show()

```

```python
# Install required packages

import pandas as pd
import numpy as np
import yfinance as yf
import datetime
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow import keras
from tensorflow.keras import layers

# Define the pair of stocks (IBM and INTC)
stock_symbols = ['IBM', 'INTC']

# Fetching the data for both stocks
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime(2024, 1, 31)

stock_data = {}
for symbol in stock_symbols:
    stock_data[symbol] = yf.download(symbol, start=start_date, end=end_date, progress=False)

# Extracting the 'Close' prices for both stocks
close_prices = pd.DataFrame({symbol: data['Close'] for symbol, data in stock_data.items()})
close_prices = close_prices.dropna()

# Calculate the spread between the two stocks (IBM - INTC)
spread = close_prices['IBM'] - close_prices['INTC']

# Feature Scaling for fast training of neural networks
scaler = MinMaxScaler()
spread_scaled = scaler.fit_transform(spread.values.reshape(-1, 1))

# Convert the data into sequences for predicting the price difference after the next 5 days
def create_sequences(df, target_col_index, forecast_horizon=5):
    sequences = []
    target_values = []
    for i in range(len(df) - 30 - forecast_horizon):
        sequence = df[i: i + 30]
        target_value = df[i + 30 + forecast_horizon, target_col_index]  # Adjust the index for the target value
        sequences.append(sequence)
        target_values.append(target_value)
    return np.array(sequences), np.array(target_values)

# Assuming the target column index for spread is 0
forecast_horizon = 5
x_train, y_train = create_sequences(spread_scaled, target_col_index=0, forecast_horizon=forecast_horizon)

# Define the transformer-like model
inputs = layers.Input(shape=(30, 1))
x = layers.MultiHeadAttention(num_heads=4, key_dim=64)(inputs, inputs)
x = layers.GlobalAveragePooling1D()(x)
x = layers.Dense(64, activation="relu")(x)
outputs = layers.Dense(1, activation="linear")(x)
model = keras.Model(inputs=inputs, outputs=outputs)

# Compile the transformer-like model
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss="mean_absolute_error",
)

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32)

# Generate predictions using a rolling window for the next 5 days
rolling_predictions = []
for i in range(len(spread_scaled) - 30 - forecast_horizon):
    window = spread_scaled[i:i + 30].reshape(1, -1, 1)
    prediction = model.predict(window)
    rolling_predictions.append(prediction)

# Convert the predictions to NumPy array
rolling_predictions = np.concatenate(rolling_predictions, axis=0)

# Invert the scaling for the predicted values
predicted_spread = scaler.inverse_transform(rolling_predictions)

# Create a DataFrame for the predicted spread
predicted_df = pd.DataFrame(predicted_spread, columns=['Predicted_Spread'])

# Plotting predicted differences versus actual differences for the next 5 days
actual_spread = spread.values[30 + forecast_horizon:]  # Extract actual spread for the same period

plt.figure(figsize=(10, 6))
plt.plot(actual_spread, label='Actual Spread', color='blue')
plt.plot(predicted_df['Predicted_Spread'], label='Predicted Spread', color='red', linestyle='dashed')
plt.title(f'Predicted vs Actual Spread (Next {forecast_horizon} Days)')
plt.xlabel('Time Steps')
plt.ylabel('Spread')
plt.legend()
plt.show()

# Calculate mean and variance of spread differences for the next 5 days
spread_diff = actual_spread - predicted_df['Predicted_Spread'].values
mean_diff = np.mean(spread_diff)
variance_diff = np.var(spread_diff)

# Print mean and variance of spread differences for the next 5 days
print(f"Mean of Actual Minus Predicted Spread Differences (Next {forecast_horizon} Days): {mean_diff}")
print(f"Variance of Actual Minus Predicted Spread Differences (Next {forecast_horizon} Days): {variance_diff}")

# Pairs Trading Strategy
threshold = 0.5  # Adjust this threshold based on your analysis

# Create signals (1 for long, -1 for short, 0 for hold)
signals = np.where(spread_diff > threshold, -1, np.where(spread_diff < -threshold, 1, 0))

# Implied Trades based on signals
implied_trades = np.diff(signals)

# Calculate the daily returns based on the implied trades
daily_returns = implied_trades * actual_spread[:-1]

# Cumulative returns
cumulative_returns = np.cumsum(daily_returns)

# Calculate Maximum Drawdown
running_max = np.maximum.accumulate(cumulative_returns)
drawdown = (cumulative_returns - running_max) / running_max
max_drawdown = np.min(drawdown)

# Plot cumulative returns
plt.figure(figsize=(10, 6))
plt.plot(cumulative_returns, label='Cumulative Returns', color='green')
plt.title('Pairs Trading Cumulative Returns')
plt.xlabel('Time Steps')
plt.ylabel('Cumulative Returns')
plt.legend()
plt.show()

# Print Maximum Drawdown
print(f"Maximum Drawdown: {max_drawdown:.6f}")

# Back-testing Performance Measures
total_trades = np.count_nonzero(implied_trades)
positive_trades = np.count_nonzero(implied_trades > 0)
negative_trades = np.count_nonzero(implied_trades < 0)

winning_percentage = (positive_trades / total_trades) * 100 if total_trades > 0 else 0

average_daily_return = np.mean(daily_returns)
std_daily_return = np.std(daily_returns)

sharpe_ratio = (average_daily_return / std_daily_return) * np.sqrt(252)  # Assuming 252 trading days in a year

# Print back-testing performance measures
print(f"Total Trades: {total_trades}")
print(f"Positive Trades: {positive_trades}")
print(f"Negative Trades: {negative_trades}")
print(f"Winning Percentage: {winning_percentage:.2f}%")
print(f"Average Daily Return: {average_daily_return:.6f}")
print(f"Standard Deviation of Daily Return: {std_daily_return:.6f}")
print(f"Sharpe Ratio: {sharpe_ratio:.6f}")
```
